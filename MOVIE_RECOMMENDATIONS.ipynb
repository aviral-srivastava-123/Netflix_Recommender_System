```python
new_meta = meta.loc[meta.year == 2017,['genres','id','title','year']]
new_meta.head()
```

```python
new_meta['id'] = new_meta['id'].astype(int)
```

```python
#merge both the data
data2 = pd.merge(new_meta, credits, on ='id')
data2.head()
```

```python
data2['genres'].dtypes
```

```python
data2['genres'] = data2['genres'].map(lambda x: ast.literal_eval(x))
data2['cast'] = data2['cast'].map(lambda x: ast.literal_eval(x))
data2['crew'] = data2['crew'].map(lambda x: ast.literal_eval(x))
data2['genres'].dtypes
```

```python
data2['genres']
```

```python
# take out all the genres form the column
def make_genreslist(x):
    gen = []
    s=""
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
        if gen == []:
            return np.Nan
        else:
            return (s.join(gen))
```

```python
data2['genres_list'] = data2['genres'].map(lambda x: make_genreslist(x))
data2['genres_list']
```

```python
def get_actor_1(x):
    casts=[]
    for i in x:
        casts.append(i.get('name'))
    if casts==[]:
        return np.NaN
    else:
        return casts[0]
```

```python
data2['actor_1_name']= data2['cast'].map(lambda x : get_actor_1(x))
data2['actor_1_name']
```

```python
def get_actor_2(x):
    casts=[]
    for i in x:
        casts.append(i.get('name'))
    if casts==[] or len(casts)<=1:
        return np.NaN
    else:
        return casts[1]
```

```python
data2['actor_2_name']=data2['cast'].map(lambda x:get_actor_2(x))
```

```python
def get_actor_3(x):
    casts=[]
    for i in x:
        casts.append(i.get('name'))
    if casts==[] or len(casts)<=2:
        return np.NaN
    else:
        return casts[2]
```

```python
data2['actor_3_name']=data2['cast'].map(lambda x:get_actor_3(x))
```

```python
def get_directors(x):
    dt = []
    st=""
    for i in x:
        if i.get('job')=='Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
data2['director_name']=data2['crew'].map(lambda x: get_directors(x))
```
